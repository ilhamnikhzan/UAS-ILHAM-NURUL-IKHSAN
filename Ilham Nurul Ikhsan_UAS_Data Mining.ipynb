# K-Means Clustering dengan Python

Pada notebook ini, kita akan menggunakan algoritma K-Means untuk melakukan klastering data. Algoritma ini akan membantu kita mengelompokkan data ke dalam beberapa cluster berdasarkan kemiripan.

Berikut adalah langkah-langkah yang akan kita lakukan:
1. Mengimpor pustaka yang diperlukan.
2. Menghasilkan data sintetis.
3. Memvisualisasikan data.
4. Standarisasi data.
5. Menerapkan algoritma K-Means.
6. Memvisualisasikan hasil kastering.


## Langkah 1: Mengimpor Pustaka yang Diperlukan

Kita akan mengimpor beberapa pustaka Python yang diperlukan untuk klastering data menggunakan K-Mean.


```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
```

## Langkah 2: Menghasilkan Data Sintetis

Kita akan menggunakan `make_blobs` dari `sklearn.datasets` untuk membuat data sintetis yang terdiri dari beberapa klaste.


```python
# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Convert to DataFrame
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])
```

## Langkah 3: Memvisualisasikan Data

Mari kita lihat bagaimana data kita terlihat sebelum klasterin.


```python
# Visualize the data
plt.figure(figsize=(10, 6))
plt.scatter(df['Feature 1'], df['Feature 2'], s=50)
plt.title('Synthetic Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Langkah 4: Standarisasi Data

Agar K-Means bekerja dengan baik, kita perlu menstandarisasi data agar memiliki mean 0 dan standard deviation .


```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

## Langkah 5: Menerapkan Algoritma K-Means

Sekarang kita akan menerapkan algoritma K-Means dengan 4 cluste.


```python
# Apply K-Means clustering with explicit n_init
kmeans = KMeans(n_clusters=4, random_state=0, n_init=10)
kmeans.fit(X_scaled)
```

## Langkah 6: Memvisualisasikan Hasil Klastering

Mari kita lihat hasil klastering dengan memvisualisasikan data yang telah dikelompokkan ke dalam 4 cluste.


```python
# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Feature 1', y='Feature 2', hue='Cluster', palette='viridis', s=50)
plt.title('K-Means Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Langkah Opsional: Memvisualisasikan Centroid

Kita juga dapat memvisualisasikan centroid dari setiap cluster untuk melihat pusat dari masing-masing cluste.


```python
# Optional: Plot the centroids
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Feature 1', y='Feature 2', hue='Cluster', palette='viridis', s=50, legend=False)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroids')
plt.title('K-Means Clustering with Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
